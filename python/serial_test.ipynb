```python
import serial
import csv
import pandas as pd
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
s = serial.Serial('/dev/tty.usbmodemfa131', 115200)
print(s.readline())
```

```python
data = []
n = 100
with open('data.csv', 'w') as f:
    writer = csv.writer(f)
    for i in range(n):
        for j in range(n):
            s.write(bytearray("W {:d} {:d};".format(i, j), 'ascii'))
            ain = int(s.readline())
            row = [i, j, ain]
            writer.writerow(row)
            data.append(row)
data = pd.DataFrame(np.array(data), columns=['DAC0', 'DAC1', 'AIN'])
```
